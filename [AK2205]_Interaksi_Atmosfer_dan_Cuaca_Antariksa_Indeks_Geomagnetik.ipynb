## Tugas Interaksi Atmosfer dan Cuaca Antariksa

```python
#impor modul pandas
import pandas as pd

#read the data
data = pd.read_csv('/content/ourdata.xlsx - mydata.csv')

#drop beberapa kolom yang ngga digunakan
data.drop(["h", "y","mmdd","d"],axis=1, inplace=True)
```

## Informasi dalam data

```python
data.head
```

```python
#ubah format data pada kolom 'date' menjadi datetime
data["date"]
data["date"] = pd.to_datetime(data["date"])

date = data["date"]
```

## Plotting keseluruhan plot untuk melihat "dinamika" masing-masing index

```python
#plotting
import matplotlib.pyplot as plt
plt.title("Kp")
plt.plot(data["date"], data["kp"])
plt.xlabel("date")
plt.ylabel("(index)*10")
plt.xticks(rotation=45)
plt.show()

plt.title("R (Sunspot Number)")
plt.plot(data["date"], data["R"], color='red')
plt.xlabel("date")
plt.ylabel("sunspot no")
plt.xticks(rotation=45)
plt.show()

plt.title("DST-Index")
plt.plot(data["date"], data["DST-Index"], color='purple')
plt.xlabel("date")
plt.ylabel("sfu ($10^{-22}Wm^{-2}Hz^{-1}$)")
plt.xticks(rotation=45)
plt.show()

plt.title("f10.7_index")
plt.plot(data["date"], data["f10.7_index"], color='green')
plt.xlabel("date")
plt.ylabel("nT")
plt.xticks(rotation=45)
plt.show()

plt.title("AE-index")
plt.plot(data["date"], data["AE-index"], color='cyan')
plt.xlabel("date")
plt.ylabel("nT")
plt.xticks(rotation=45)
plt.show()

plt.title("AL-index")
plt.plot(data["date"], data["AL-index"], color='grey')
plt.xlabel("date")
plt.ylabel("nT")
plt.xticks(rotation=45)
plt.show()

plt.title("AU-index")
plt.plot(data["date"], data["AU-index"], color='orange')
plt.xlabel("date")
plt.ylabel("nT")
plt.xticks(rotation=45)
plt.show()
```
